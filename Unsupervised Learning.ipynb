```python
# ---טעינת הדאטה---  
# בקטע הקוד הזה איחדתי 80 קבצי דאטה לקובץ אחד, שזה כולל 20 עונות ו4 הליגות של אנגליה 
# התעזרתי עם הצאט בקטע הקוד הזה נתתי לו את הרעיון שלי כתבתי לו כמה טיפים  
# כתבתי לו קוד קצר אמרתי לו שיעשה לולאה ויאחד אותם ביחד ולמחוק את עמודת הזמן כי זאת עמודה בעייתית 

import pandas as pd

leagues = ["E0", "E1", "E2", "E3"]  # הליגות של אנגליה ה4:פרמיירליג, צ׳מפיונשיפ, ליגה 1, ליגה 2
num_seasons = 20   #מ2006 עד 2026

all_dfs = []

for i in range(num_seasons):        #לולאה על מספר העונות  
    for league in leagues:          #לולאה על הליגות 

        
        # אם זו העונה הראשונה – שם הקובץ בלי מספר בסוגריים
        if i == 0:
            file_name = f"C:/Users/2023/Downloads/{league}.csv"  
        # אחרת – שם הקובץ כולל מספר עונה בסוגריים
        else:
            file_name = f"C:/Users/2023/Downloads/{league} ({i}).csv"      
 
        print("טוען קובץ:", file_name)

        # הכנסת הקובץ תוך פאנדאס
        df = pd.read_csv(file_name)

  #רוב הקבצים לא כוללים את עמודת Time, ולכן חשוב למחוק אותה כדי שכל הקבצים יהיו באותו מבנה ואפשר יהיה לאחד אותם בלי שגיאות
  # מחיקת עמודת Time אם קיימת
        if "Time" in df.columns:
            df = df.drop(columns=["Time"])


        all_dfs.append(df)


# איחוד כל הדאטה
df = pd.concat(all_dfs, ignore_index=True)               
# שמירה לקובץ
output_file = "england_all_cleaned.csv"
df.to_csv(output_file, index=False)

print("הקובץ המאוחד נשמר:", output_file) # נוודא אם הקובץ נשמר בהצלחהנ
print("סה״כ שורות:", len(df))        # בדיקה כמה יש לנו שורות 
```

```python
# זיהוי המשחק:
# Season     - עונת המשחק (למשל 2020/2021)
# League     - שם הליגה (Premier League, Championship וכו')
# Date       - תאריך המשחק
# HomeTeam   - הקבוצה הביתית
# AwayTeam   - הקבוצה האורחת

# תוצאה סופית:
# FTHG       - Full Time Home Goals (שערים של הקבוצה הביתית בסיום)
# FTAG       - Full Time Away Goals (שערים של הקבוצה האורחת בסיום)
# FTR        - Full Time Result (תוצאה כוללת: H=ניצחון ביתי, D=תיקו, A=ניצחון חוץ)

# תוצאת מחצית:
# HTHG       - Half Time Home Goals (שערים ביתיים במחצית)
# HTAG       - Half Time Away Goals (שערים אורחים במחצית)
# HTR        - Half Time Result (תוצאת המחצית: H, D, A)

# סטטיסטיקה התקפית:
# HS         - Home Shots (בעיטות לשער של הקבוצה הביתית)
# AS         - Away Shots
# HST        - Home Shots on Target (בעיטות למסגרת ביתיות)
# AST        - Away Shots on Target (בעיטות למסגרת אורחות)

# עבירות ומשמעת:
# HF         - Home Fouls (עבירות שביצעה הקבוצה הביתית)
# AF         - Away Fouls (עבירות שביצעה הקבוצה האורחת)
# HC         - Home Corners (קרנות של הקבוצה הביתית)
# AC         - Away Corners (קרנות של הקבוצה האורחת)
# HY         - Home Yellow Cards (צהובים לקבוצה הביתית)
# AY         - Away Yellow Cards (צהובים לקבוצה האורחת)
# HR         - Home Red Cards (אדומים לקבוצה הביתית)
# AR         - Away Red Cards (אדומים לקבוצה האורחת) 
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt  
from sklearn.metrics import silhouette_score
```

```python
# מוחקים את כל העמודות שיש בתוכם סטרינג
new_df= df.drop(columns=['Date','HomeTeam', 'AwayTeam','FTR','HTR'])
```

```python
# --- נרמול כל העמודות המספריות ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(new_df)
```

```python
#silhouette_score
scores = []

for k in range(2, 20):
    model = KMeans(n_clusters=k).fit(X_scaled)
    score = silhouette_score(X_scaled, model.labels_)
    scores.append(score)
    print(f"k={k}, silhouette={score:.4f}")

print("הציונים:", scores)
```

```python
# ציור גרף של ה(silhouette)
plt.plot(range(2, 20), scores, marker='o')
plt.title('Silhouette Scores')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()
```

```python
k=2 Silhouette התוצאה הכי טובה שקיבלתי  
קיבל ציון הכי גבוה 0.0959
```

```python
# KMeans 
res = KMeans(n_clusters=2, random_state=42)
labels = res.fit_predict(X_scaled)
```

```python
labels_predicted=res.labels_
```

```python
# ציר x בעיטות בית
# ציר y  (עבירות שביצעה הקבוצה הביתית)
plt.scatter(new_df['HS'], new_df['HF'], c=labels_predicted,s=1)
plt.xlabel("HS")
plt.ylabel("HF")
plt.title("KMeans Clustering (Based on All Numeric Features)")
plt.show()
```
